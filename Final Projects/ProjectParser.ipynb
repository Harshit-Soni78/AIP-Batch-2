<!-- hay gpt!

I have a CSV file that contains these columns : "Timestamp,Email Address,Name,Project Title,A short Description of Project,Have you made a README.md file for your project ?,Github Repo Link for Project"

I want wo make a folder named final_projects, and in this folder I want everybody's repos downloaded and unzipped,

folder structure should look like,

```
final_projects\
            | name(project title)\
                                | rest project files files from github
            
            | name(project title)\
                                | rest project files files from github
            
            | name(project title)\
                                | rest project files files from github
            
            | name(project title)\
                                | rest project files from github
            
            README.md {this readme should contain submitor name, project title and project description nicely formated}
```

Please write a python script to do this task -->

````python
import os
import csv
import git
import shutil
from pathlib import Path
````

````python

def sanitize_filename(name):
    return "".join(c for c in name if c.isalnum() or c in (" ", "_", "-")).rstrip()

def download_repos(csv_file):
    final_dir = ""
    os.makedirs(final_dir, exist_ok=True)
    
    readme_content = "# Project Submissions\n\n"
    
    with open(csv_file, newline='', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        
        for row in reader:
            name = sanitize_filename(row['Name'].strip())
            project_title = sanitize_filename(row['Project Title'].strip())
            description = row['A short Description of Project'].strip()
            repo_link = row['Github Repo Link for Project'].strip()
            
            if not repo_link:
                print(f"Skipping {name} ({project_title}) - No repo link provided.")
                continue
            
            project_folder = os.path.join(final_dir, f"{name} ({project_title})")
            os.makedirs(project_folder, exist_ok=True)
            
            print(f"Cloning {repo_link} into {project_folder}...")
            try:
                git.Repo.clone_from(repo_link, project_folder)
            except Exception as e:
                print(f"Failed to clone {repo_link}: {e}")
                continue
            
            readme_content += f"## {name} - {project_title}\n\n"
            readme_content += f"**Description:** {description}\n\n"
    
    readme_path = os.path.join(final_dir, "README.md")
    with open(readme_path, "w", encoding='utf-8') as f:
        f.write(readme_content)
    
    print("All repositories processed successfully!")

````

````python

# Run the script with your CSV file
download_repos("submissions.csv")
````
